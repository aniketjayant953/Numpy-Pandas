```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-versicolor':1,'Iris-virginica':2,'Iris-setosa':0})
iris.sample(5)
```

```python
# cmap and aplha
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='summer',alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Plot size

```python
plt.figure(figsize=(10,5))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='summer',alpha=0.9)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Annotations

```python
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df.head()
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(len(sample_df)):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'red'})
```

## Subplots

```python
# A diff way to plot graph
batters.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike rate')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike rate')
fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0].set_title('Avg vs Strike Rate')
ax[0].set_ylabel('Strike rate')

ax[1].scatter(batters['avg'],batters['runs'])
ax[1].set_title('Avg vs Runs')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')

fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'],color='yellow')
ax[1,0].hist(batters['avg'],color='green')
ax[1,1].hist(batters['runs'],color='red')

fig.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
ax1.scatter(batters['avg'],batters['runs'],color='yellow')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(223)
ax3.hist(batters['avg'])
```

## 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
yy.shape
```

```python
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

## Contour Plots

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

## Heatmap

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run'] == 6)]
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),np.arange(1,21))
plt.xticks([0,1,2,3,4,5],list(range(1,7)))
plt.colorbar()
```
