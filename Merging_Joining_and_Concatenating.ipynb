```python
import pandas as pd
import numpy as np
```

```python
courses = pd.read_csv('courses.csv')
students = pd.read_csv('students.csv')
nov = pd.read_csv('reg-month1.csv')
dec = pd.read_csv('reg-month2.csv')

matches = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,27]
})
temp_df
students = pd.concat([students,temp_df],ignore_index='True')
```

```python
# find total revenue generated
regs.merge(courses,on='course_id')['price'].sum()
```

```python
#2. find month by month revenue
temp_df = pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
temp_df.merge(courses,on='course_id').groupby('level_0')['price'].sum()
```

```python
# 3. print the registration table
# cols -> name -> courses -> price   
regs.merge(students,on='student_id').merge(courses,on='course_id')[['name','course_name','price']]
```

```python
# plot bar chart for revenue/courses
regs.merge(courses,on='course_id').groupby('course_name')['price'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
# 5. find students who enrolled in both the months
common_stu_id = np.intersect1d(nov['student_id'],dec['student_id'])
common_stu_id
```

```python
students[students['student_id'].isin(common_stu_id)]
```

```python
# 6. find course that got no enrollment
# courses['course_id']
# regs['course_id']

not_common = np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(not_common)]
```

```python
# 7. find students who did not enroll into any courses
students_list = np.setdiff1d(students['student_id'],regs['student_id'])
students[students['student_id'].isin(students_list)]

# (10/28)*100
```

```python
# 8. Print student name-> partner name for all enrolled students
students.merge(students,left_on='partner',right_on='student_id',how='inner')[['name_x','name_y']]
```

```python
# 9. find top 3 students who did not most number enrollment
regs.merge(students,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending = False)
```

```python
# 10. find top 3 students who spent most amount of money on courses
regs.merge(students,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).reset_index()
```

```python
temp_df = delivery.merge(matches,left_on='match_id',right_on='id')
temp_df.head(2)
```

```python
six_df = temp_df[temp_df['batsman_runs'] == 6]
```

```python
# stadium -> sixes
num_sixes = six_df.groupby('venue')['venue'].count()
num_matches = matches['venue'].value_counts()
```

```python
(num_sixes/num_matches).sort_values(ascending=False).head(3)
```

```python
# find orange cap holder of all the seasons
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep='first').sort_values('season')
```
