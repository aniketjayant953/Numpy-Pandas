```python
from textblob import TextBlob

text = 'hi I can drve at nigt'
TextBlob("Haven't").correct().string
```

```python
# punctuation 
import string 
string.punctuation
```

```python
def remove_punctuation(text):
    
    for i in string.punctuation:
        if i in text:
            text = text.replace(i,'')
    return text
```

```python
remove_punctuation('Hi! How are you?')
```

```python
df['review'] = df['review'].apply(remove_punctuation)
```

```python
from nltk.tokenize import word_tokenize
```

```python
df['tokens'] = df['review'].apply(word_tokenize)
```

```python
# stopwords removal
from nltk.corpus import stopwords
len(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    l = []
    for word in text:
        if word not in stopwords.words('english'):
            l.append(word)
    return l
```

```python
df['tokens'] = df['tokens'].apply(remove_stopwords)
```

```python
df['review'] = df['tokens'].apply(lambda x: ' '.join(x))
```

```python
df['char_len'] = df['review'].str.len()
```

```python
df['words_len'] = df['tokens'].apply(len)
```

```python
# BoW

from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(max_features=5000,ngram_range=(1,3))
bag_of_words = count_vectorizer.fit_transform(df['review'])
bag_of_words = pd.DataFrame(bag_of_words.toarray(),columns = count_vectorizer.get_feature_names())
```

```python
bag_of_words
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(bag_of_words.values)
```
