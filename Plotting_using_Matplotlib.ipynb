## 2D line plot



- Bivariate Analysis

# Scatter Plots



- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-1,10,50)
y = 10*x+3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='+',s=100)
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Avgerge')
plt.ylabel('Strike Rate')
```

```python
# size
# slower
tips = sns.load_dataset('tips')
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

### plt.plot vs plt.scatter

```python
# scatterplot using plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
```

# Bar chart



- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['Red','Blue','Green','Yellow','Pink']

plt.bar(colors,children,color='black')
```

```python
# horizontal bar chart using data
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
# problem is this is overlapping
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
# multiple Bar Chart
# xticks
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='yellow',label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue',label='2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='red',label='2017')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.legend()
plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['Red is color that is loved by children','Blue is color that is loved by children','Green is color that is loved by children','Yellow is color that is loved by children','Pink is color that is loved by children']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Stacked Bar Chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015'] + df['2016']),label='2017')
plt.legend()
plt.show()
```

# Pie Chart



- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
marks = ['english','science','maths','sst','hindi']
plt.pie(data,labels=marks)
plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','yellow','pink','brown','red'])
plt.show()
```

```python
# explode, shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```
